Text Data TF-IDF Vectorization with scikit-learn

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Combining the selected columns into a single text column
data['combined_features'] = data['Cuisines'] + " " + data['Rating color'] + " " + data['Rating text'] + " " + data['Votes'].astype(str)
# Create the TF-IDF vectorizer and fit it to your text data
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(data['combined_features'])

#Transforming the text documents using the trained vectorizer
x = tfidf_vectorizer.transform(data['combined_features'])
```

Calculating Cosine Similarity Score with sklearn

```python
from sklearn.metrics.pairwise import cosine_similarity
similarityScore = cosine_similarity(x)
similarityScore
```

```python
# User Input for Preferred Cuisine
userInput = input("Enter your preferred cuisine: ")

# Finding the indices of restaurants with the preferred cuisine
indices = [i for i, cuisines in enumerate(data["Cuisines"]) if userInput in cuisines]

# Calculating Recommendation Scores for the preferred cuisine
RecommendationScore = list(enumerate(similarityScore[indices[0]]))

# Sorting the restaurant list by Recommendation Score
sortedRestaurantList = sorted(RecommendationScore, key=lambda x: x[1], reverse=True)

# Printoing the top recommended restaurants
count = 0
for rest in sortedRestaurantList:
    Index = rest[0]
    k = data.iloc[Index]["Restaurant Name"]
    print(k)
    count += 1
    if count == 10:
        break
```
